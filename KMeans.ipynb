```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Prepare the Dataset
# Simulating a dataset of customer purchase history
data = {
    'CustomerID': range(1, 11),  # 10 customers
    'Annual_Spend': [500, 1200, 800, 1300, 700, 300, 1600, 1000, 900, 1100],  # in dollars
    'Frequency_of_Purchases': [5, 10, 8, 12, 7, 3, 15, 9, 7, 11]  # Number of purchases per year
}

# Create a DataFrame
df = pd.DataFrame(data)

# Step 2: Preprocess the Data (Scaling)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['Annual_Spend', 'Frequency_of_Purchases']])

# Step 3: Applying K-Means Clustering
# Set number of clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_data)

# Step 4: Visualize the Results
# Scatter plot for visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=df['Annual_Spend'],
    y=df['Frequency_of_Purchases'],
    hue=df['Cluster'],
    palette='Set2',
    s=100
)
plt.title('Customer Segmentation Based on Purchase History')
plt.xlabel('Annual Spend (in USD)')
plt.ylabel('Frequency of Purchases')
plt.show()

# Step 5: Display the Clusters
print("Customer Groups:")
print(df[['CustomerID', 'Annual_Spend', 'Frequency_of_Purchases', 'Cluster']])
```
